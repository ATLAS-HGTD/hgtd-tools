# Convert Slot Table csv to json

this is done to minimize load on the API infrastructure, given that the slot table itself is just a constant dump of all positions in different coordinate systems along with their static information

```python
import pandas as pd
```

```python
csv_file = pd.DataFrame(pd.read_csv("Slot_Table_fullJune2025_demo54mod.csv", sep = ",", header = 0, index_col = False))
```

```python
csv_file
```

```python
csv_file.to_json("Slot_fullJune2025_demo54mod.json", orient = "records", force_ascii = True, default_handler = None)
```
