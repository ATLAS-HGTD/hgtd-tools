# Convert Slot Table csv to json

this is done to minimize load on the API infrastructure, given that the slot table itself is just a constant dump of all positions in different coordinate systems along with their static information.


To reproduce:

download the Slots from https://nginx-hgtddb.app.cern.ch/partslist

and the Slot attributes from https://nginx-hgtddb.app.cern.ch/listattributes  
(keep only these columns: `part_serial_number,FT_Length_mm,FT_length_category,DU_F_or_B,Layer,Module,Used_for_NTC_Interlock,PEB_type,Quadrant,Row,SU_Module,SU_Row,SU_type,Disk_per_vessel,Disk_side_from_IP,Vessel`)

saving both as csv.

The csv needs to be prepared to only include `,` (no `;`) and no trailing spaces in the header row.

Then you can read them in again and export as json, mimicking a successful API GET request.

```python
import pandas as pd
```

```python
fields = ['part_serial_number','FT_Length_mm','FT_length_category','DU_F_or_B','Layer','Module','Used_for_NTC_Interlock','PEB_type','Quadrant','Row','SU_Module','SU_Row','SU_type','Disk_per_vessel','Disk_side_from_IP','Vessel']
```

```python
csv_file = pd.DataFrame(pd.read_csv("Slot_Table_fullJuly2026_withPreliminaryNTC.csv", sep = ",", header = 0, index_col = False, usecols=fields))
```

```python
csv_file
```

```python
csv_file.to_json("Slot_Table_fullJuly2026_withPreliminaryNTC.json", orient = "records", force_ascii = True, default_handler = None)
```

```python
csv_file_ = pd.DataFrame(pd.read_csv("Slot_fullJuly2026_withPreliminaryNTC.csv", sep = ",", header = 0, index_col = False))
```

```python
csv_file_
```

```python
csv_file_.to_json("Slot_fullJuly2026_withPreliminaryNTC.json", orient = "records", force_ascii = True, default_handler = None)
```
